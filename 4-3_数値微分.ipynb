### 4.3.1 微分

```python
# 数値微分の悪い実装例
def numerical_diff(f, x):
    h = 10e-50
    return (f(x+h) - f(x)) / h

# 数値微分良い実装例
def numerical_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)
```

### 4.3.2 数値微分の例

$y = 0.01x ^ 2 + 0.1x$ をPythonで実装する

```python
def function_1(x):
    return 0.01*x**2 + 0.1*x
```

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.arange(0.0, 20.0, 0.1) # 0から20まで、0.1刻みのx配列
```

```python
y = function_1(x)
```

```python
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
```

```python
# x=5のときの微分
numerical_diff(function_1, 5)
```

```python
# x=10のときの微分
numerical_diff(function_1, 10)
```

```python
# 微分した傾きをプロットする
def tangent_line(f, x):
    d = numerical_diff(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y
```

```python
tf = tangent_line(function_1, 5)
y2 = tf(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
tf = tangent_line(function_1, 10)
y2 = tf(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

### 4.3.3 偏微分

引数の2乗和の計算  
$f(x0, x1) = x0^2 + x1^2$


```python
def funtion_2(x):
    return np.sum(x**2)
    # または return np.sum(x**2)
```

x0=3, x1=4のときのx0に対する偏微分 $ \frac{∂u}{∂x0}$ を求める

```python
def function_tmp1(x0):
    return x0*x0 + 4.0**2.0
```

```python
numerical_diff(function_tmp1, 3.0)
```

x0=3, x1=4のときのx0に対する偏微分 $ \frac{∂u}{∂x1}$ を求める

```python
def function_tmp2(x1):
    return 3.0**2.0 + x1*x1
```

```python
numerical_diff(function_tmp2, 4.0)
```
